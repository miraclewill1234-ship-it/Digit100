```python
import spacy
from collections import Counter
import pygal
import re


nlp = spacy.load("en_core_web_sm")


with open("2x1.xml", "r", encoding="utf-8") as file:
    text = file.read()

clean_text = re.sub(r"<.*?>", "", text)


doc = nlp(clean_text)


verbs = []
for token in doc:
    if token.pos_ == "VERB":
        verbs.append(token.lemma_)


counts = Counter(verbs)


with open("verbFreq.txt", "w") as o:
    for verb, count in counts.most_common():
        o.write(f"{verb}: {count}\n")


chart = pygal.Bar()
chart.title = "Top 10 Verbs in Arthur XML"

for verb, count in counts.most_common(10):
    chart.add(verb, count)

chart.render_to_file("verb_chart.svg")

print("Done")
```
